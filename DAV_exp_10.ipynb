```python
%%html
<!DOCTYPE html>
<html>
<head>
  <title>D3.js Mouse Events</title>
  <script src="https://d3js.org/d3.v6.min.js"></script>
  <style>
    .circle {
      fill: steelblue;
      stroke: black;
      stroke-width: 1px;
    }
  </style>
</head>
<body>
  <h1>D3.js Mouse Events</h1>
  <script>
    const width = 500;
    const height = 300;

    const svg = d3.select("body")
      .append("svg")
      .attr("width", width)
      .attr("height", height);

    const circleGroup = svg.append("g");

    const data = [
      { cx: 100, cy: 100, radius: 50 },
      { cx: 300, cy: 200, radius: 30 },
      { cx: 400, cy: 100, radius: 40 }
    ];

    const circles = circleGroup.selectAll("circle")
      .data(data)
      .enter()
      .append("circle")
      .attr("class", "circle")
      .attr("cx", d => d.cx)
      .attr("cy", d => d.cy)
      .attr("r", d => d.radius)
      .on("mouseover", function() {
        d3.select(this)
          .transition()
          .duration(200)
          .attr("r", d => d.radius * 1.2);
      })
      .on("mouseout", function() {
        d3.select(this)
          .transition()
          .duration(200)
          .attr("r", d => d.radius);
      })
      .on("click", function() {
        d3.select(this)
          .transition()
          .duration(200)
          .style("fill", "orange");
      });
  </script>
</body>
</html>

```

```python
%%html
<!DOCTYPE html>
<html>
<head>
  <title>D3.js Drag Events</title>
  <script src="https://d3js.org/d3.v6.min.js"></script>
  <style>
    .circle {
      fill: steelblue;
      stroke: black;
      stroke-width: 1px;
      cursor: move;
    }
  </style>
</head>
<body>
  <h1>D3.js Drag Events</h1>
  <script>
    const width = 500;
    const height = 300;

    const svg = d3.select("body")
      .append("svg")
      .attr("width", width)
      .attr("height", height);

    const circleGroup = svg.append("g");

    const data = [
      { cx: 100, cy: 100, radius: 50 },
      { cx: 300, cy: 200, radius: 30 },
      { cx: 400, cy: 100, radius: 40 }
    ];

    const circles = circleGroup.selectAll("circle")
      .data(data)
      .enter()
      .append("circle")
      .attr("class", "circle")
      .attr("cx", d => d.cx)
      .attr("cy", d => d.cy)
      .attr("r", d => d.radius)
      .call(d3.drag()
        .on("start", dragStarted)
        .on("drag", dragged)
        .on("end", dragEnded));

    function dragStarted(event, d) {
      d3.select(this)
        .raise()
        .classed("active", true);
    }

    function dragged(event, d) {
      d3.select(this)
        .attr("cx", d.cx = event.x)
        .attr("cy", d.cy = event.y);
    }

    function dragEnded(event, d) {
      d3.select(this)
        .classed("active", false);
    }
  </script>
</body>
</html>
```
